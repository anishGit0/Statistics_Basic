1. Types of Data
    1. Qualitative (Categorical) Data:

      Describes qualities, categories, or characteristics.

      Cannot be measured numerically but can be classified.

      Examples:

      Colors of cars (Red, Blue, Black)

      Customer feedback (Good, Average, Bad)

      Gender (Male, Female)

      Types of Qualitative Data:

      Nominal: Categories without any order.

      Example: Eye color (Blue, Brown, Green)

      Ordinal: Categories with a meaningful order but no fixed difference.

      Example: Customer satisfaction levels (Poor, Fair, Good, Excellent)

      2. Quantitative (Numerical) Data:

      Represents numerical values that can be measured or counted.

      Examples:

      Height of students (in cm)

      Number of mobile phones sold

      Temperature in Celsius

      Types of Quantitative Data:

      Interval: Numeric data with equal intervals but no true zero.

      Example: Temperature in Celsius (0°C does not mean "no temperature")

      Ratio: Numeric data with equal intervals and a true zero.

      Example: Weight in kg (0 kg means no weight)












2. Measures of Central Tendency:

  These are statistical measures that summarize a dataset by identifying a central value.

    1. Mean (Average):

  Formula: Sum of all values ÷ Number of values

  Example: The average score of students in a class.

  When to Use:

  When data has no extreme values (outliers).

  Best for normally distributed data.

  Example Situation: Calculating the average salary of employees in a company.


    2. Median (Middle Value):

  The middle value when data is arranged in order.

  Example: If test scores are 50, 60, 70, 80, 90, the median is 70.

  When to Use:

  When data has outliers (e.g., income data where a few very high salaries can skew the mean).

  Example Situation: Finding the median house price in a city to avoid skewed results from luxury properties.

    3. Mode (Most Frequent Value):

      The value that appears most frequently in a dataset.

      Example: In {2, 3, 3, 4, 5, 3}, the mode is 3.

      When to Use:

      When identifying the most common category or preference.

      Example Situation: Finding the most popular mobile phone brand based on customer purchases.

3. Dispersion

  Dispersion refers to the spread or variability of data points within a dataset. It indicates how much the values deviate from a central value, typically the mean.

  Variance

  Variance is a measure of dispersion that quantifies the average squared deviation of each data point from the mean. A higher variance indicates greater data spread.

  Calculation: It's calculated by summing the squared differences between each data point and the mean, and then dividing by the number of data points (or n-1 for a sample).

  Standard Deviation

  Standard deviation is the square root of the variance. It provides a more interpretable measure of dispersion in the original units of the data.

  Interpretation: A higher standard deviation indicates greater variability, while a lower standard deviation indicates less variability and more data clustering around the mean.

  In essence, variance and standard deviation are closely related and both quantify the spread of data around the mean. The standard deviation is often preferred because it's easier to understand and relate to the original data.

4. Box Plot (Box-and-Whisker Plot):

  A box plot is a graphical representation of data distribution that shows minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values.

  What It Shows:

  Median (Q2): The middle value of the dataset.

  Interquartile Range (IQR = Q3 - Q1): Measures data spread between Q1 and Q3.

  Whiskers: Extend to the smallest and largest values within 1.5 × IQR.

  Outliers: Points outside the whiskers indicate extreme values.

  Insights from a Box Plot:

  Skewness:

  Longer upper whisker → Right-skewed (positive skew).

  Longer lower whisker → Left-skewed (negative skew).

  Spread of Data:

  A wider box means more variability.

  Outliers:

  Identifies extreme values that may need further investigation.

5. Role of Random Sampling in Inference:

  Random sampling is a method of selecting a subset from a population where each member has an equal chance of being chosen. It helps in making accurate and unbiased inferences about the entire population.

  Why It’s Important:

  Reduces Bias – Ensures the sample represents the population fairly.

  Improves Accuracy – Leads to reliable conclusions with minimal errors.

  Enables Generalization – Findings from the sample can be applied to the whole population.

6. Concept of Skewness:

  Skewness measures the asymmetry of a dataset’s distribution. It shows whether data is symmetrically distributed or leans toward one side.

  Types of Skewness:

  Positive Skew (Right-Skewed)

  Tail extends to the right (higher values).

  Mean > Median > Mode

  Example: Income distribution (few very high salaries).

  Negative Skew (Left-Skewed):

  Tail extends to the left (lower values).

  Mean < Median < Mode

  Example: Exam scores (most students score high, few fail).

  No Skew (Symmetrical Distribution):

  Mean = Median = Mode

  Example: Heights of people in a population.

  Effect on Data Interpretation:

  Skewness impacts measures of central tendency (mean, median, mode).

  In highly skewed data, the median is a better measure of center than the mean.

  Outliers influence skewness, affecting data analysis and decision-making.

7. Interquartile Range (IQR):

  The IQR measures the spread of the middle 50% of data and helps identify outliers.

  Formula:

  IQR = Q3 − Q1

  Where:

  Q1 (First Quartile): 25th percentile

  Q3 (Third Quartile): 75th percentile

  How IQR Detects Outliers:

  Lower Bound:

  Q1 − 1.5 × IQR

  Upper Bound:

  Q3 + 1.5 × IQR

  Outliers: Any data points outside these bounds are considered outliers.

8. Conditions for Using Binomial Distribution
The binomial distribution applies when an experiment meets the following conditions:

  Fixed Number of Trials (n): The experiment is repeated a set number of times.

  Only Two Outcomes: Each trial results in success or failure (e.g., heads or tails).

  Constant Probability (p): The probability of success remains the same for each trial.

  Independent Trials: The outcome of one trial does not affect others.

  Example Use Cases:

  Tossing a coin 10 times and counting heads.

  Checking if a batch of products has defects.

  Probability of a student passing a test based on preparation.

9. Properties of Normal Distribution

  Bell-Shaped & Symmetrical: The mean, median, and mode are equal.

  Mean-Centered: Most values cluster around the mean.

  Asymptotic: The tails extend infinitely but never touch the x-axis.

  Defined by Mean (μ) & Standard Deviation (σ): Determines the shape and spread.

  Empirical Rule (68-95-99.7 Rule)

  For a normal distribution:

  68% of data falls within 1σ of the mean.

  95% falls within 2σ.

  99.7% falls within 3σ.

  Use Cases:
  
  Analyzing test scores in a class.

  Predicting stock price variations.

  Measuring IQ score distribution.

10. Let's consider the number of customers arriving at a coffee shop during the morning rush hour (between 8:00 AM and 9:00 AM). Assume that, on average, 30 customers arrive during this hour. This can be modeled as a Poisson process.

  Probability Calculation:

  Question: What is the probability that exactly 35 customers will arrive between 8:00 AM and 9:00 AM?

  Formula:

  The probability mass function for a Poisson distribution is given by:

  P(X = k) = (λ^k * e^(-λ)) / k!

  where:

  P(X = k) is the probability of observing exactly k events
  
  λ is the average rate of events (in this case, 30 customers per hour)
  
  e is the base of the natural logarithm (approximately 2.71828)
  
  k! is the factorial of k (e.g., 5! = 5 * 4 * 3 * 2 * 1)

  Calculation:

  In this case, we want to find P(X = 35) with λ = 30. Plugging these values into the formula:

  P(X = 35) = (30^35 * e^(-30)) / 35!

11. Random Variable:

  A random variable is a numerical outcome of a random experiment. It assigns values to each possible outcome.

  Types of Random Variables:
  
  Discrete Random Variable:

  Takes countable values (e.g., 0, 1, 2, …).

  Example: Number of heads in 5 coin flips.

  Continuous Random Variable:

  Takes infinite values within a range.

  Example: Height of students (e.g., 160.5 cm, 162.3 cm).

  Key Difference:
  
  Discrete → Finite or countable values (e.g., number of customers).

  Continuous → Measurable values over an interval (e.g., weight, temperature).

  Covariance

  Covariance measures the directional relationship between two variables.

  Formula: Cov(X, Y) = Σ[(Xi - X̄)(Yi - Ȳ)] / (n - 1) where:

  Xi, Yi are individual data points
  
  X̄, Ȳ are the means of X and Y
  
  n is the number of data points
  
  Using this formula on our dataset you can calculate the covariance between 'Hours Studied' and 'Exam Score'.

  Correlation

  Correlation measures the strength and direction of the linear relationship between two variables.

  Formula: Corr(X, Y) = Cov(X, Y) / (σX * σY)
  where:
  
  Cov(X, Y) is the covariance
  
  σX, σY are the standard deviations of X and Y
  
  We can calculate the correlation between 'Hours Studied' and 'Exam Score' using the computed covariance and the standard deviations of each variable.

  Interpretation:

  Covariance: A positive covariance suggests that as 'Hours Studied' increases, 'Exam Score' tends to increase as well. The magnitude of the covariance isn't easily interpretable in terms of relationship strength.

  Correlation: A positive correlation close to +1 indicates a strong positive linear relationship. This means that as 'Hours Studied' increases, 'Exam Score' is very likely to increase as well. A correlation of +1 would indicate a perfect positive linear relationship.
